```python
import pandas as pd
import matplotlib.pyplot as plt

# Importing data and filtering only movies
netflix_df = pd.read_csv(r"C:\Users\user\Desktop\Netflix.csv", index_col="index")
movies_only = netflix_df[netflix_df['type'] == 'Movie']

# Calculation of average movie duration by release year
duration_trends = movies_only.groupby('release_year')['duration'].mean()

# Trend visualization
duration_trends.plot(figsize=(10, 5), color = "red", linewidth=2)

# Styling the chart
plt.title("Evolution of average movie length on Netflix (1940-2021)", fontsize=14, fontweight="bold")
plt.xlabel("Release year", fontsize=12)
plt.ylabel("Average duration (min)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.4)

# Result output 
plt.show()
```
